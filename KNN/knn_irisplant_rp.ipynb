### 2.- Separar el conjunto en características (X) y clases (y)

```python
X = ds[["x1","x2"]]
y = ds[["variety"]]
```

### 3.- Validación cruzada

```python
from sklearn.model_selection import train_test_split
```

### 3.1.- Proporción 80/20 

80% de los datos para entrenamiento (X_train,y_train)

20% para prueba (X_test,y_test)

```python
X80_train,X20_test,y80_train,y20_test = train_test_split(X,y,test_size=0.20)
```

### 3.2 Proporción 70/30

### 4.- Intanciar clasificador KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sbs
import matplotlib.pyplot as plt
```

### 4.1.- KNN con 1 vecino equivalente a clasificador Mínimas Distancias

### 4.1.1.- KNN 1 vecino distancia Manhattan

```python
def entrenar_clf(model,X_train,X_test,y_train,y_test,tittle):
    model.fit(X_train,y_train.values.ravel())
    model_predicted = model.predict(X_test)
    confusion_ma = confusion_matrix(y_test,model_predicted)
    print("Predicción del modelo:")
    print(model_predicted)
    print("Actual:")
    print(y_test.values.ravel())
    #calculando la presicion (accuracy) del clasificador
    accuracy = model.score(X_test,y_test.values.ravel())
    print('accuracy',accuracy*100,'%')
    print('----------- \n')
    print("Matriz de confusión")
    w=["Setosa","Versicolor",'Virginica']
    sbs.heatmap(data=confusion_ma,cmap="Pastel1",annot=True,xticklabels=w,yticklabels=w)
    plt.xlabel("Actual")
    plt.ylabel("predicción")
    plt.title(tittle)
```

```python
knn_1M = KNeighborsClassifier(n_neighbors=1,p=1)
print("proporción 80/20")
entrenar_clf(model=knn_1M,X_train=X80_train,X_test=X20_test,y_train=y80_train,y_test=y20_test,tittle="proporción 80/20")
```

```python
print("proporción 70/30")
entrenar_clf(model=knn_1M,X_train=X70_train,X_test=X30_test,y_train=y70_train,y_test=y30_test,tittle="proporción 70/30")
```

```python
print("proporción 60/40")
entrenar_clf(model=knn_1M,X_train=X60_train,X_test=X40_test,y_train=y60_train,y_test=y40_test,tittle="proporción 60/40")
```

```python
from sklearn.metrics import roc_curve,auc
```

### 5.4.- Función para graficar curva ROC

```python
def plot_roc_curve(model,X_test,y_test):
    model.fit(X_test,y_test.values.ravel())
    y_scores = model.predict_proba(X_test)
    fpr, tpr, threshold = roc_curve(y_test, y_scores[:, 1])
    roc_auc = auc(fpr,tpr)
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('Verdadero Positivo')
    plt.xlabel('Falso Positivo')
    plt.show()
```

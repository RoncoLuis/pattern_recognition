```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
```

```python
fertility_ds = pd.read_csv("../datasets/fertility/fertility.csv")
fertility_ds["diagnosis"] = fertility_ds["diagnosis"].replace(["N"],0) #normal
fertility_ds["diagnosis"] = fertility_ds["diagnosis"].replace(["O"],1) #altered
```

```python
data = fertility_ds.columns.tolist()[:-1]
target = fertility_ds.columns.tolist()[-1]
X = fertility_ds[data]
y = fertility_ds[target]
fertility_ds.head()
```

```python
#Matriz de correlación
X.corr()
```

```python
sbs.heatmap(X.corr(),xticklabels=X.columns.values,yticklabels=X.columns.values)
plt.title("Matriz de correlación fertility")
plt.show()
```

```python
pd.plotting.scatter_matrix(X,figsize=(20,12))
plt.show()
```

```python
import pandas as pd
```

```python
ds = pd.read_csv("../datasets/breast-cancer-wisconsin/PCA_breast-cancer-wisconsin.csv")
```

```python
ds.head()
```

```python
import matplotlib.pyplot as plt
```

```python
maligno = ds[ds["diagnosis"]=='M']
benigno = ds[ds["diagnosis"]=='B']
```

```python
plt.scatter(x=maligno["x1"],y=maligno["x2"],color='r',marker='o')
plt.scatter(x=benigno["x1"],y=benigno["x2"],color='g',marker='^')
plt.legend(["Maligno","Benigno"])
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Distribución 2 componentes")
```

```python
principal = ds[ds["diagnosis"]=='M']
nube = ds[ds["diagnosis"]=='B']
mean_principal = principal.mean()
pMedio = (mean_principal+mean_nube)/2
```

```python
plt.scatter(x=principal["x1"],y=principal["x2"],color='r',marker='o')
plt.scatter(x=nube["x1"],y=nube["x2"],color='g',marker='^')
plt.scatter(x=mean_principal[0],y=mean_principal[1],color='b',marker='o',s=100)
plt.scatter(x=mean_nube[0],y=mean_nube[1],color='k',marker='^',s=100)
plt.scatter(x=pMedio[0],y=pMedio[1],color='y',marker='*',s=300)
plt.legend(["Principal","Nube","mean_P","mean_N","pMedio"])

plt.plot([-1000,4000], [-250,800],'m--')

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Distribución 2 componentes breast-cancer-wisconsin")
```

```python
import pandas as pd
```

```python
ds = pd.read_csv("../datasets/iris_plant/PCA_iris_plant.csv")
```

```python
ds.head()
```

```python
import matplotlib.pyplot as plt
```

```python
setosa = ds[ds["variety"]==0]
versicolor = ds[ds["variety"]==1]
virginica = ds[ds["variety"]==2]
mean_versi = versicolor.mean()
mean_virg = virginica.mean()
```

```python
mean_nube=(mean_versi+mean_virg)/2
```

```python
plt.scatter(x=setosa["x1"],y=setosa["x2"],color='r',marker='o')
plt.scatter(x=virginica["x1"],y=virginica["x2"],color='g',marker='^')
plt.scatter(x=versicolor["x1"],y=versicolor["x2"],color='b',marker='s')
#plt.scatter(x=mean_nube[0],y=mean_nube[1],color='k',marker='s',s=100)
plt.legend(["Setosa","Versicolor","Virginica"])
plt.xlabel("sepal.length")
plt.ylabel("sepal.width")
plt.title("Distribución 2 componentes")
```

```python
ds["variety"] = ds["variety"].replace([2],1)
```

```python
principal = ds[ds["variety"]==1]
nube = ds[ds["variety"]==0]
```

```python
plt.scatter(x=principal["x1"],y=principal["x2"],color='r',marker='o')
plt.scatter(x=nube["x1"],y=nube["x2"],color='g',marker='^')
plt.legend(["Principal","Nube"])
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Distribución 2 componentes")
```

```python
mean_principal = principal.mean()
mean_nube = nube.mean()
print(mean_principal)
```

```python
plt.scatter(x=principal["x1"],y=principal["x2"],color='r',marker='o')
plt.scatter(x=nube["x1"],y=nube["x2"],color='g',marker='^')
plt.scatter(x=mean_principal[0],y=mean_principal[1],color='b',marker='o',s=200)
plt.scatter(x=mean_nube[0],y=mean_nube[1],color='k',marker='^',s=200)
plt.legend(["Principal","Nube","mean_P","mean_N"])
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Distribución 2 componentes")
```

```python
pMedio = (mean_principal+mean_nube)/2
print(pMedio)
```

```python
plt.scatter(x=principal["x1"],y=principal["x2"],color='r',marker='o')
plt.scatter(x=nube["x1"],y=nube["x2"],color='g',marker='^')
plt.scatter(x=mean_principal[0],y=mean_principal[1],color='b',marker='o',s=200)
plt.scatter(x=mean_nube[0],y=mean_nube[1],color='k',marker='^',s=200)
plt.scatter(x=pMedio[0],y=pMedio[1],color='y',marker='*',s=300)
plt.legend(["Principal","Nube","mean_P","mean_N","pMedio"])
plt.plot([-3,4], [-0.7,1.5],'m--')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Distribución 2 componentes iris_plant")
plt.
```
